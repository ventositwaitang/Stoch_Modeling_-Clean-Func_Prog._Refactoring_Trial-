## Binomial tree pricing
 Using the formula of 
 $S_{\frac {2k}{n}}=100e^{\frac {0.16k}{n}}+0.3{\sqrt \frac  {2}{n}}Z_k$
,  obtaining graphs with path 
simulated following:

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

image.png

```python
y_k = [1,-1]

def path_generate(n):
    z_k = [None] * n   # [None] <- a list with one element which contains the special object called None. That one element is None, no value; [] <- empty list, zero values
    s_price = [None] * n #
    for k in range(0,n):
        if k == 0:
            z_k[k] = random.choice(y_k)
            s_price[k] = 100
        else:
            z_k[k]=z_k[k-1] + random.choice(y_k)
            s_price[k] = 100*np.exp(0.16*k/n + 0.3*np.sqrt(2/n)*z_k[k])
    return s_price
```

*   Simulation Paths using Binomial Tree with 5 steps:


```python
#Number of Simulation paths
paths = 5 

#Number of Steps sims
for n in [5]:#,50,200,1000]:
    #loop Number of Simulation paths we want
    for i in range(1,paths+1):
        s_price = path_generate(n)
        plt.plot(list(range(1,n+1)), s_price)
    plt.title('Simulation Paths using Binomial Tree with ' +str(n) +' steps')
    plt.ylabel('Stock Price')
    plt.xlabel('Number of steps')
    #filename = '../contents/ML_skillset/Graph for '+ str(n) + '.pdf'
    #plt.savefig(filename)
    #plt.figure().clear('all')
```

```python
# importing required modules
import matplotlib.pyplot as plt
import numpy as np
 
# function to generate coordinates
def create_plot(ptype):
    # setting the x-axis values
    x = np.arange(-10, 10, 0.01)
     
    # setting the y-axis values
    if ptype == 'linear':
        y = x
    elif ptype == 'quadratic':
        y = x**2
    elif ptype == 'cubic':
        y = x**3
    elif ptype == 'quartic':
        y = x**4
             
    return(x, y)
 
# setting a style to use
plt.style.use('fivethirtyeight')
 
# create a figure
fig = plt.figure()
 
# define subplots and their positions in figure
plt1 = fig.add_subplot(221)
plt2 = fig.add_subplot(222)
plt3 = fig.add_subplot(223)
plt4 = fig.add_subplot(224)
 
# plotting points on each subplot
x, y = create_plot('linear')
plt1.plot(x, y, color ='r')
plt1.set_title('$y_1 = x$')
 
x, y = create_plot('quadratic')
plt2.plot(x, y, color ='b')
plt2.set_title('$y_2 = x^2$')
 
x, y = create_plot('cubic')
plt3.plot(x, y, color ='g')
plt3.set_title('$y_3 = x^3$')
 
x, y = create_plot('quartic')
plt4.plot(x, y, color ='k')
plt4.set_title('$y_4 = x^4$')
 
# adjusting space between subplots
fig.subplots_adjust(hspace=.5,wspace=0.5)
 
# function to show the plot
plt.show()
```

```python
import numpy as np
import scipy.stats as st

def callprice(sigma):
    d1 = (np.log(S/K)+(r-delta+0.5*sigma**2)*T)/(sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    N_d1 = st.norm(0,1).cdf(d1)
    N_d2 = st.norm(0,1).cdf(d2)
    
    c = S * np.exp(-delta*T)*N_d1 - K * np.exp(-r*T)*N_d2
    return c

def diff(sigma):
    return callprice(sigma) - c_given

```

```python
def imp_sigma():
    # Manually setting upper and lower bound
    lowerb = 0.000001
    upperb = 1

    # Set up initial value of volatility
    sigma = (lowerb + upperb)/2

    while abs(callprice(sigma)-c_given) > 0.00000000001:
        # if sigma and lowerb are on the same side, set sigma as upper bound
        if diff(sigma)*diff(lowerb)<0:
            upperb=sigma
            sigma = (lowerb+upperb)/2

        #if sigma and lowerb are not on the same side, set sigma as lower bound
        else:
            lowerb=sigma
            sigma = (lowerb+upperb)/2
    else:
        print('implied volatility (' + str(K) + ') = ' + str(sigma))
```

```python
import numpy as np
import scipy.stats as st
from random import random

import matplotlib.pyplot as plt
```

```python
def maximum():
    # Generate a stock price path with N = 1000 steps
    lnS_T = np.log(S) + np.cumsum((r-delta-0.5*sigma**2)*h + sigma*np.sqrt(h) * \
                                st.norm.ppf(np.random.uniform(0,1,size=N)))
    S_T = np.exp(lnS_T)
    # Obtain stock price path starting from 0.5 to 1 and get the maximum
    S_T_calc = S_T[int(N/2):]
    maximum = max(S_T_calc)
#For visualization
    if visual_switch == 0:
        pass
    else:
        plt.plot(list(np.arange(0,1,h)),S_T)
    return maximum
```

```python
def main_run():
    call_payoff = [None] * num_sim
    for i in range(0,num_sim):
        call_payoff[i] = max(maximum()-K,0)
    tot_call_payoff=np.math.fsum(call_payoff)
    c = tot_call_payoff * np.exp(-r*T)/num_sim
    print('Option Price with ' + str(num_sim) + ' simulations is ' + str(c))
```
